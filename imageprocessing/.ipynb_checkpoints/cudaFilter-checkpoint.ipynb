```python
import numpy as np
import pycuda.driver as cuda
import pycuda.autoinit
from pycuda.compiler import SourceModule
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from timeit import default_timer as timer

%matplotlib inline
#%qtconsole --style monokai
```

```python
mod = SourceModule("""
    __global__ void divide(float *d_out, float *d_in){
        int idx = blockDim.x*blockIdx.x + threadIdx.x;
        float f = d_in[idx];
        d_out[idx] =  f/2 ;
    }
    
    __global__ void filterReduceBW(float *d_out, float *d_in){
        int idx_in  = blockIdx.x*blockDim.x*3 + threadIdx.x*3;
        int idx_out = blockIdx.x*blockDim.x + threadIdx.x;
        float f = d_in[idx_in]*0.299 + d_in[idx_in+1]*0.587 +d_in[idx_in+2]*0.114;
        d_out[idx_out]= f ;
    }
    
    __global__ void filterSquareBlur(float *d_out, float *d_in){
        int idx = blockIdx.x*blockDim.x*3 + threadIdx.x*3 ;
        const int N = 25;
        int stencil[N] =   {idx-2*blockDim.x*3-2*3,idx-2*blockDim.x*3-1*3,idx-2*blockDim.x*3,idx-2*blockDim.x*3+1*3,idx-2*blockDim.x*3+2*3,
                            idx-1*blockDim.x*3-2*3,idx-1*blockDim.x*3-1*3,idx-1*blockDim.x*3,idx-1*blockDim.x*3+1*3,idx-1*blockDim.x*3+2*3,
                            idx-0*blockDim.x*3-2*3,idx-0*blockDim.x*3-1*3,idx-0*blockDim.x*3,idx-0*blockDim.x*3+1*3,idx-0*blockDim.x*3+2*3,
                            idx+1*blockDim.x*3-2*3,idx+1*blockDim.x*3-1*3,idx+1*blockDim.x*3,idx+1*blockDim.x*3+1*3,idx+1*blockDim.x*3+2*3,
                            idx+2*blockDim.x*3-2*3,idx+2*blockDim.x*3-1*3,idx+2*blockDim.x*3,idx+2*blockDim.x*3+1*3,idx+2*blockDim.x*3+2*3
                            }; 
        float rgb[3];
        for(int i=0;i<N;i++){
            int idx_temp = stencil[i];
            for (int icolor=0; icolor<3;icolor++){
                rgb[icolor] += d_in[ idx_temp+icolor ];
                }
            }
        d_out[idx]   = rgb[0]/N;
        d_out[idx+1] = rgb[1]/N;
        d_out[idx+2] = rgb[2]/N;
    }
    """)
square = mod.get_function("divide")
filterReduceBW   = mod.get_function("filterReduceBW")
filterSquareBlur = mod.get_function("filterSquareBlur")
```

# 0. Show Orignal Image

```python
plt.imshow(mpimg.imread('im.jpg'))
```

# 1. Reduce Black White

```python
h_in  = mpimg.imread('im.jpg').astype(np.float32)
h_out = np.empty_like(h_in[:,:,0])
d_in  = cuda.mem_alloc(h_in.nbytes)
d_out = cuda.mem_alloc(h_out.nbytes)

cuda.memcpy_htod(d_in,h_in)
NROW,NCOL = h_in.shape[0] , h_in.shape[1]

########################################
# 1. GPU performance
start = timer()
filterReduceBW(d_out,d_in,grid=(NROW,1,1),block=(NCOL,1,1))
end = timer()
print("GPU time: {:3.3} ms".format(1000*(end - start)))
########################################

########################################
# 2. CPU performance
start = timer()
h_in = np.dot(h_in,np.array([0.299,0.587,0.114]))
end = timer()
print("CPU time: {:3.3} ms".format(1000*(end - start))) 
########################################


cuda.memcpy_dtoh(h_out,d_out)
plt.imshow(h_out,cmap='viridis')
d_out.free()
d_in.free()
del h_in,h_out
```

# 2. Square Blur

```python
h_in  = mpimg.imread('im.jpg').astype(np.float32)
h_out = np.empty_like(h_in)
d_in  = cuda.mem_alloc(h_in.nbytes)
d_out = cuda.mem_alloc(h_out.nbytes)

cuda.memcpy_htod(d_in,h_in)
NROW,NCOL = h_in.shape[0] , h_in.shape[1]

########################################
# GPU performance
start = timer()
filterSquareBlur(d_out,d_in,grid=(NROW,1,1),block=(NCOL,1,1))
end = timer()
print("GPU time: {:3.3} ms".format(1000*(end - start)))
########################################

cuda.memcpy_dtoh(h_out,d_out)
d_out.free()
d_in.free()
h_out = h_out[2:-2,2:-2,:]
plt.imshow(h_out/255)
plt.savefig("blur.jpg")
```
